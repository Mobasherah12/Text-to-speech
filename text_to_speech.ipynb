```python
from gtts import gTTS

text_to_say = "This is a sample piece of text read by GTTS."

language = "en"

gtts_object = gTTS(text = text_to_say, 
                  lang = language,
                  slow = False)

gtts_object.save("/content/gtts.wav")
```

```python
from IPython.display import Audio

Audio("/content/gtts.wav")
```

```python
english_text = " You can launch a new career in web development today by learning HTML & CSS. You don't need a computer science degree or expensive software. All you need is a computer, a bit of time, a lot of determination, and a teacher you trust. I've taught HTML and CSS to countless coworkers and held training sessions for fortune 100 companies. I am that teacher you can trust. "

english_language = "en"

english_gtts_object = gTTS(text = english_text,
                          lang = english_language,
                          slow = True)

english_gtts_object.save("/content/english.wav")

Audio("/content/english.wav")
```
